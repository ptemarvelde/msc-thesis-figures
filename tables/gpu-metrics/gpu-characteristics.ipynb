```python
from util import *
import pandas as pd
df = pd.read_csv("./tables/gpu-metrics/gpu-metrics.tsv", delimiter="\t")
df = df.set_index(["Group", "Characteristic", "Unit"])
df
```

```python
table = df.to_latex(
    index=True,
    na_rep="",
)
table = (
    table.replace("{lllllllll}", "{p{0.11\linewidth}p{0.12\linewidth}p{0.08\linewidth}rrrrrr}")
    .replace("{*}", "{\linewidth}")
    .replace("\midrule", "\midrule\midrule")
    .replace("\cline{1-9} \cline{2-9}\n\\bottomrule", "\\bottomrule")
    .replace(
        "Characteristic",
        "Character-\\newline istic$\downarrow$",
    ).replace("Pascal", "Pas.")
    .replace("Volta", "Vol.")
    .replace("Turing", "Tur.")
    .replace("Ampere", "Amp.")
    .replace("&  & P100", "GPU $\\rightarrow$ & & P100")
    .replace("Group", "Group $\downarrow$")
    .replace("Unit", "Unit $\downarrow$")
    .replace("Characteristic", "Characteristic $\downarrow$")
)
save_table(table, 'appendices/auto-generated/gpu-characteristics.tex')
```
