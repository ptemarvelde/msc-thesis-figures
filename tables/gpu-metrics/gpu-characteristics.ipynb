```python
from figure_utils import *
import pandas as pd
df = pd.read_csv("./tables/gpu-metrics/gpu-metrics.tsv", delimiter="\t")
df = df.set_index(["Group", "Unit", "Characteristic",])
```

```python
# Replace commas and convert to numeric
transformed = df.droplevel(1).reset_index()
new_columns = (transformed['Group'].fillna("") + " " + transformed['Characteristic']).str.lower().str.replace(" ", "_")

numbers = transformed.drop(columns=['Group', 'Characteristic']).replace(',', '', regex=True).apply(pd.to_numeric, errors='ignore').T
numbers.columns = new_columns
display(numbers)
```

```python
table = df.to_latex(
    index=True,
    na_rep="",
)
# display(df)
table = (
    table.replace("{llllllllll}", "{p{0.11\linewidth}p{0.08\linewidth}p{0.12\linewidth}rrrrrrr}")
    .replace("{*}", "{\linewidth}")
    .replace("\midrule", "\midrule\midrule")
    .replace("\cline{1-9} \cline{2-9}\n\\bottomrule", "\\bottomrule")
    .replace(
        "Characteristic",
        "Character-\\newline istic$\downarrow$",
    ).replace("Pascal", "Pas.")
    .replace("Volta", "Vol.")
    .replace("Turing", "Tur.")
    .replace("Ampere", "Amp.")
    .replace("&  & P100", "GPU $\\rightarrow$ & & P100")
    .replace("Group", "Group $\downarrow$")
    .replace("Unit", "Unit $\downarrow$")
    .replace("Characteristic", "Characteristic $\downarrow$")
)
# print(table)
save_table(table, 'appendices/auto-generated/gpu-characteristics.tex')
```
